```python
from matplotlib import pylab as plt
import numpy as np
```

```python
def eg2():
    """
    The equation for a straight line is y=mx+c, m is the gradient, c is the y-intercep
    """
    x = np.linspace(start=-1.,stop=10.,num=1000)
    m, c = 9., 10.
    y = m*x+c
    fig, ax = plt.subplots()
    ax.plot(y,x)
    ax.set_xlabel('x (time)')
    ax.set_ylabel('y (distance)')
    ax.set_title("A straight line graph")
    ax.set_xlim(0.,)
    return ( fig, ax, x, y, m, c)
```

```python
 fig, ax, x, y, m, c = eg2()
```

```python
fig.savefig("straightLine",format="pdf")
```
